```python
import pandas as pd
```

```python
df = pd.read_csv('matches.csv', parse_dates=['date'], low_memory=False)
```

```python
df
```

```python
df[(((df.home_team == 'Legia Warszawa')&(df.away_team == 'Lech Poznań')) | ((df.home_team == 'Lech Poznań')&(df.away_team == 'Legia Warszawa'))) & (df.date.dt.year >= 2010) & (df.date.dt.year <= 2015)]
```

```python
df['goal_diff'] = abs(df['home_team_goal'] - df['away_team_goal'])
```

```python
df
```

```python
biggest_goal_diffs = df.sort_values('goal_diff', ascending=False).head(10)
```

```python
biggest_goal_diffs
```

```python
df['goals_per_match'] = df['home_team_goal'] + df['away_team_goal']
```

```python
leagues = df.groupby('league').agg({'goals_per_match': 'mean'})
```

```python
leagues
```

```python
leagues['goals_per_match'].sort_index().plot(kind='bar')
```
